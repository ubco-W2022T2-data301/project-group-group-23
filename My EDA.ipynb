```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/processed/player_per_game_cleaned.csv')
players = ['Stephen Curry', 'Ray Allen', 'Reggie Miller']
df_players = df[df['player_name'].isin(players)]
df_players = df_players[['player_name', 'season', 'points_per_game', 'three_point_percentage']]
```

```python
df_players.boxplot(column='points_per_game', by='player_name')
plt.title('Distribution of Points per Game by Player')
plt.ylabel('Points per Game')
plt.xlabel('Player')
plt.show()
```

```python
colors = {'Stephen Curry':'red', 'Ray Allen':'green', 'Reggie Miller':'blue'}
plt.scatter(df_players['three_point_percentage'], df_players['points_per_game'], c=df_players['player_name'].apply(lambda x: colors[x]))
plt.title('Scatterplot of Points per Game vs. Three-Point Percentage')
plt.xlabel('Three-Point Percentage')
plt.ylabel('Points per Game')
plt.legend(colors)
plt.show()
```

```python
df_players.pivot_table(index='season', columns='player_name', values='points_per_game').plot()
plt.title('Points per Game by Season')
plt.xlabel('Season')
plt.ylabel('Points per Game')
plt.show()
```

Based on the boxplot, we can see that Stephen Curry has the highest median and range of points per game, followed by Ray Allen and then Reggie Miller.

In the scatterplot, we see a positive correlation between three point percentage and points per game for all three players, with Stephen Curry having the highest correlation.

In the line chart, we see that Stephen Curry's scoring has been increasing over the years despite some up and downs, while Ray Allen's scoring has remained relatively constant until last few years of his carrer, and Reggie Miller's scoring has decrease at a stable rate after his peak.

Overall, these visualizations suggest that Stephen Curry has been the most dominant scorer of the three players, with a strong correlation between his shooting accuracy and scoring, and a trend of increasing scoring over time.
