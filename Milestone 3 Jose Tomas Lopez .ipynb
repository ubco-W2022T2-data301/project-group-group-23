```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="ticks", font_scale=1.3)
plt.rc("axes.spines", top=False, right=False)
```

```python
df = pd.read_csv("data/processed/Player Shooting Processed.csv", index_col=[0])
df
```

```python
violin= sns.scatterplot(data=df, x='season', y='percent_assisted_two_point_field_goal',)
df=df.loc[(df['percent_assisted_two_point_field_goal']<1) & (df['percent_assisted_two_point_field_goal']>0)]
violin.set(xlabel='Year/Season', ylabel='% Assisted Two Point Field Goal', title='Percent of Two Point Field Goals')
```

```python
sns.scatterplot(data=df, x='season', y='field_goal_percentage',)
```

```python
sns.boxplot(x="season", y="field_goal_percentage", data=df,
            whis=[0, 100], width=.6, palette="vlag")
```

```python
sns.scatterplot(data=df, x='season', y='percent_fga_from_x2p_range',)
```

```python
sns.boxplot(x="season", y="percent_fga_from_x2p_range", data=df,
            whis=[0, 100], width=.6, palette="vlag")
```
